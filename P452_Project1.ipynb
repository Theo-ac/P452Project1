```python
import qiskit
from qiskit_aer import AerSimulator
from qiskit_aer.primitives import SamplerV2
```

```python
n_qubits = 10
def measure_Circuit(circuit):
    sampler = SamplerV2()
    job = sampler.run([circuit], shots=128)
    result_ideal = job.result()
    counts_ideal = result_ideal[0].data.meas.get_counts()
    print('Counts(ideal):', counts_ideal)
    return circuit

```

```python
example = qiskit.QuantumCircuit(n_qubits)
example.h(0)
example.cx(0, 1)
example.cx(1, 2)
example.measure_all()
histogram = measure_Circuit(example)
```
